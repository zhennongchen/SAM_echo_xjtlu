```python
import sys
sys.path.append('/host/d/Github')
import Whole_heart_segmentation_junzhe.functions_collection as ff
import os
```

```python
data_path = '/host/d/GitHub/Whole_heart_segmentation_junzhe/example_data/data'
```

```python
case_list = ff.find_all_target_files(['*0002'],data_path)
print(case_list)
```

```python
# create patient (data) list
results = []
for i in range(0,case_list.shape[0]):
    case = case_list[i]
    print(case)

    case_id = os.path.basename(case)

    # find out how many slices in this case
    img_folder = os.path.join(case,'img')
    slice_list = ff.sort_timeframe(ff.find_all_target_files(['*.nii.gz'],img_folder), 2, '_', '.')
    num_slices = slice_list.shape[0]


    for j in range(0,num_slices):
        slice_file = slice_list[j]
        slice_index = ff.find_timeframe(slice_file, 2, '_', '.')
        
        # image_file:
        image_file = slice_file

        # corresponding label file
        seg_folder = os.path.join(case,'seg')
        label_file = os.path.join(seg_folder,'slice_'+str(slice_index)+'.nii.gz')
        
        results.append([case_id, slice_index, image_file, label_file])

# make results as as pandas dataframe
import pandas as pd
df = pd.DataFrame(results, columns=['case_id','slice_index','img_file','seg_file'])
# add one column called batch, evenly split the data into 2 batches
df['batch'] = pd.qcut(df.index, 2, labels=False)

# save to excel
output_excel = os.path.join(os.path.dirname(data_path),'Patient_list/patient_list.xlsx')
df.to_excel(output_excel, index=False)
print('Patient list saved to: ', output_excel)

    
```
